```python
train_data =[]
train_label = []
```

```python
files = glob.glob("./train/*.txt")
for file_name in files:
    # Get Lables
    basename = os.path.basename(file_name)
    language = basename.split("-")[0]
    # Export text
    file = open(file_name, "r", encoding="utf-8")
    text = file.read()
    text = text.lower()
    file.close
    # Get frequency of each letter appearance
    code_a = ord('a')
    code_z = ord('z')
    count = [0 for n in range(0,26)]
    for letter in text:
        code_x = ord(letter)
        if code_a <= code_x <= code_z:
            count[code_x - code_a] += 1
    # Normalizing
    total = sum(count)
    freq = list(map(lambda n: n / total, count))
    # Insert into created list
    train_data.append(freq)
    train_label.append(language)
    
```

```python
test_data =[]
test_label = []
```

```python
files = glob.glob("./test/*.txt")
for file_name in files:
    # Get Lables
    basename = os.path.basename(file_name)
    language = basename.split("-")[0]
    # Export text
    file = open(file_name, "r", encoding="utf-8")
    text = file.read()
    text = text.lower()
    file.close
    # Get frequency of each letter appearance
    code_a = ord('a')
    code_z = ord('z')
    count = [0 for n in range(0,26)]
    for letter in text:
        code_x = ord(letter)
        if code_a <= code_x <= code_z:
            count[code_x - code_a] += 1
    # Normalizing
    total = sum(count)
    freq = list(map(lambda n: n / total, count))
    # Insert into created list
    test_data.append(freq)
    test_label.append(language)
```

```python
len(test_data)
```

```python
len(test_label)
```

```python
sum(test_data[0])
```

```python
# traning
clf = svm.SVC()
clf.fit(train_data, train_label)
```

```python
# Predict
predict = clf.predict(test_data)
```

```python
# result analysis
score = metrics.accuracy_score(test_label, predict)
report = metrics.classification_report(test_label, predict)
print("Score: ", score)
print("-------------------------Report-------------------------")
print(report)
```
